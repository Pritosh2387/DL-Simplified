# Section 1: Swin Transformer (`swin_tiny_patch4_window7_224`)
The **Swin Transformer** utilizes shifted window attention to scale down self-attention computation complexity to linear, making it highly effective at capturing fine-grained facial expression patterns.

```python
# ============================================
# Emotion-Based Stress Classification
# Model: Swin Transformer
# Dataset: FER2013 (Folder-Based)
# Platform: Kaggle Notebook
# Framework: PyTorch
# ============================================

# =========================
# Install Required Packages
# =========================
!pip install timm -q

# =========================
# Import Libraries
# =========================
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from sklearn.metrics import classification_report, confusion_matrix

import torch
import torch.nn as nn
from torch.utils.data import DataLoader

import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder

import timm

from tqdm import tqdm

# =========================
# Device Configuration
# =========================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Device:", device)

# =========================
# Dataset Paths
# =========================

TRAIN_DIR = "/kaggle/input/datasets/msambare/fer2013/train"
TEST_DIR = "/kaggle/input/datasets/msambare/fer2013/test"

# =========================
# Image Transformations
# =========================

train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    
    transforms.RandomHorizontalFlip(),
    
    transforms.RandomRotation(10),
    
    transforms.ColorJitter(
        brightness=0.2,
        contrast=0.2
    ),
    
    transforms.ToTensor(),
    
    transforms.Normalize(
        mean=[0.5, 0.5, 0.5],
        std=[0.5, 0.5, 0.5]
    )
])

test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    
    transforms.ToTensor(),
    
    transforms.Normalize(
        mean=[0.5, 0.5, 0.5],
        std=[0.5, 0.5, 0.5]
    )
])

# =========================
# Load Datasets
# =========================

train_dataset = ImageFolder(
    root=TRAIN_DIR,
    transform=train_transform
)

test_dataset = ImageFolder(
    root=TEST_DIR,
    transform=test_transform
)

# =========================
# Class Names
# =========================

class_names = train_dataset.classes

print("Classes:", class_names)

print("Training Images:", len(train_dataset))
print("Testing Images:", len(test_dataset))

# =========================
# Visualize Sample Images
# =========================

plt.figure(figsize=(12, 8))

for i in range(9):

    image, label = train_dataset[i]

    image = image.permute(1, 2, 0).numpy()

    image = (image * 0.5) + 0.5

    plt.subplot(3, 3, i + 1)

    plt.imshow(image)

    plt.title(class_names[label])

    plt.axis("off")

plt.tight_layout()
plt.show()

# =========================
# Data Loaders
# =========================

BATCH_SIZE = 32

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2
)

# =========================
# Swin Transformer Model
# =========================

model = timm.create_model(
    'swin_tiny_patch4_window7_224',
    pretrained=True,
    num_classes=7
)

model = model.to(device)

# =========================
# Loss Function
# =========================

criterion = nn.CrossEntropyLoss()

# =========================
# Optimizer
# =========================

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=1e-4,
    weight_decay=1e-4
)

# =========================
# Learning Rate Scheduler
# =========================

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    patience=2,
    factor=0.5
)

# =========================
# Training Function
# =========================

def train_model(model, loader):

    model.train()

    running_loss = 0
    correct = 0
    total = 0

    for images, labels in tqdm(loader):

        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)

        correct += predicted.eq(labels).sum().item()

    epoch_loss = running_loss / len(loader)

    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc

# =========================
# Evaluation Function
# =========================

def evaluate_model(model, loader):

    model.eval()

    running_loss = 0
    correct = 0
    total = 0

    predictions = []
    actuals = []

    with torch.no_grad():

        for images, labels in tqdm(loader):

            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)

            loss = criterion(outputs, labels)

            running_loss += loss.item()

            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)

            correct += predicted.eq(labels).sum().item()

            predictions.extend(predicted.cpu().numpy())

            actuals.extend(labels.cpu().numpy())

    epoch_loss = running_loss / len(loader)

    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc, predictions, actuals

# =========================
# Training Loop
# =========================

EPOCHS = 35

train_losses = []
test_losses = []

train_accs = []
test_accs = []

best_acc = 0

for epoch in range(EPOCHS):

    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    # Training
    train_loss, train_acc = train_model(
        model,
        train_loader
    )

    # Evaluation
    test_loss, test_acc, _, _ = evaluate_model(
        model,
        test_loader
    )

    scheduler.step(test_loss)

    train_losses.append(train_loss)
    test_losses.append(test_loss)

    train_accs.append(train_acc)
    test_accs.append(test_acc)

    print(f"\nTrain Loss: {train_loss:.4f}")
    print(f"Train Accuracy: {train_acc:.2f}%")

    print(f"\nTest Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_acc:.2f}%")

    # Save Best Model
    if test_acc > best_acc:

        best_acc = test_acc

        torch.save(
            model.state_dict(),
            "best_swin_transformer_model.pth"
        )

        print("Best Model Saved!")

# =========================
# Plot Accuracy & Loss
# =========================

plt.figure(figsize=(12,5))

# Accuracy Plot
plt.subplot(1,2,1)

plt.plot(train_accs, label='Train Accuracy')

plt.plot(test_accs, label='Test Accuracy')

plt.title("Accuracy Graph")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.legend()

# Loss Plot
plt.subplot(1,2,2)

plt.plot(train_losses, label='Train Loss')

plt.plot(test_losses, label='Test Loss')

plt.title("Loss Graph")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend()

plt.tight_layout()

plt.show()

# =========================
# Load Best Model
# =========================

model.load_state_dict(
    torch.load("best_swin_transformer_model.pth")
)

# =========================
# Final Evaluation
# =========================

test_loss, test_acc, preds, labels = evaluate_model(
    model,
    test_loader
)

print(f"\nFinal Test Accuracy: {test_acc:.2f}%")

# =========================
# Classification Report
# =========================

print("\nClassification Report:\n")

print(
    classification_report(
        labels,
        preds,
        target_names=class_names
    )
)

# =========================
# Confusion Matrix
# =========================

cm = confusion_matrix(labels, preds)

plt.figure(figsize=(8,6))

plt.imshow(cm, cmap='Blues')

plt.title("Confusion Matrix")

plt.colorbar()

plt.xlabel("Predicted Labels")

plt.ylabel("True Labels")

plt.xticks(
    np.arange(len(class_names)),
    class_names,
    rotation=45
)

plt.yticks(
    np.arange(len(class_names)),
    class_names
)

plt.show()

# =========================
# Save Final Model
# =========================

torch.save(
    model.state_dict(),
    "final_swin_transformer_model.pth"
)

print("\nModel Saved Successfully!")

# =========================
# Save Class Names
# =========================

with open("class_names.txt", "w") as f:
    
    for item in class_names:
        f.write(f"{item}\n")

print("Class Names Saved!")
```

# Section 2: EfficientNetV2 (`tf_efficientnetv2_s`)
**EfficientNetV2** optimizes training speed and parameter efficiency using progressive learning and a combination of fused-MBConv and MBConv layers.

```python
# ============================================
# Emotion-Based Stress Classification
# Model: EfficientNetV2
# Dataset: FER2013 (Folder-Based)
# Platform: Kaggle Notebook
# Framework: PyTorch
# ============================================

# =========================
# Install Required Packages
# =========================
!pip install timm -q

# =========================
# Import Libraries
# =========================
import os
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import (
    classification_report,
    confusion_matrix
)

import torch
import torch.nn as nn
from torch.utils.data import DataLoader

import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder

import timm

from tqdm import tqdm

# =========================
# Device Configuration
# =========================
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print("Device:", device)

# =========================
# Dataset Paths
# =========================

TRAIN_DIR = "/kaggle/input/datasets/msambare/fer2013/train"

TEST_DIR = "/kaggle/input/datasets/msambare/fer2013/test"

# =========================
# Image Transformations
# =========================

train_transform = transforms.Compose([

    transforms.Resize((224, 224)),

    transforms.RandomHorizontalFlip(),

    transforms.RandomRotation(10),

    transforms.RandomAffine(
        degrees=10,
        translate=(0.1, 0.1)
    ),

    transforms.ColorJitter(
        brightness=0.2,
        contrast=0.2
    ),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),

    transforms.RandomErasing(
        p=0.2
    )
])

test_transform = transforms.Compose([

    transforms.Resize((224, 224)),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# =========================
# Load Datasets
# =========================

train_dataset = ImageFolder(
    root=TRAIN_DIR,
    transform=train_transform
)

test_dataset = ImageFolder(
    root=TEST_DIR,
    transform=test_transform
)

# =========================
# Class Names
# =========================

class_names = train_dataset.classes

print("Classes:", class_names)

print("Training Images:", len(train_dataset))

print("Testing Images:", len(test_dataset))

# =========================
# Visualize Sample Images
# =========================

plt.figure(figsize=(12,8))

for i in range(9):

    image, label = train_dataset[i]

    image = image.permute(1, 2, 0).numpy()

    image = np.clip(image, 0, 1)

    plt.subplot(3,3,i+1)

    plt.imshow(image)

    plt.title(class_names[label])

    plt.axis("off")

plt.tight_layout()

plt.show()

# =========================
# Data Loaders
# =========================

BATCH_SIZE = 64

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# =========================
# EfficientNetV2 Model
# =========================

model = timm.create_model(
    'tf_efficientnetv2_s',
    pretrained=True,
    num_classes=7
)

model = model.to(device)

# =========================
# Loss Function
# =========================

criterion = nn.CrossEntropyLoss(
    label_smoothing=0.1
)

# =========================
# Optimizer
# =========================

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=1e-4,
    weight_decay=1e-4
)

# =========================
# Scheduler
# =========================

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    patience=2,
    factor=0.5
)

# =========================
# Training Function
# =========================

def train_model(model, loader):

    model.train()

    running_loss = 0

    correct = 0

    total = 0

    for images, labels in tqdm(loader):

        images = images.to(device)

        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)

        correct += predicted.eq(labels).sum().item()

    epoch_loss = running_loss / len(loader)

    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc

# =========================
# Evaluation Function
# =========================

def evaluate_model(model, loader):

    model.eval()

    running_loss = 0

    correct = 0

    total = 0

    predictions = []

    actuals = []

    with torch.no_grad():

        for images, labels in tqdm(loader):

            images = images.to(device)

            labels = labels.to(device)

            outputs = model(images)

            loss = criterion(outputs, labels)

            running_loss += loss.item()

            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)

            correct += predicted.eq(labels).sum().item()

            predictions.extend(
                predicted.cpu().numpy()
            )

            actuals.extend(
                labels.cpu().numpy()
            )

    epoch_loss = running_loss / len(loader)

    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc, predictions, actuals

# =========================
# Training Loop
# =========================

EPOCHS = 30

MODEL_NAME = "efficientnetv2"

train_losses = []

test_losses = []

train_accs = []

test_accs = []

best_acc = 0

for epoch in range(EPOCHS):

    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    # Training
    train_loss, train_acc = train_model(
        model,
        train_loader
    )

    # Evaluation
    test_loss, test_acc, _, _ = evaluate_model(
        model,
        test_loader
    )

    scheduler.step(test_loss)

    train_losses.append(train_loss)

    test_losses.append(test_loss)

    train_accs.append(train_acc)

    test_accs.append(test_acc)

    print(f"\nTrain Loss: {train_loss:.4f}")

    print(f"Train Accuracy: {train_acc:.2f}%")

    print(f"\nTest Loss: {test_loss:.4f}")

    print(f"Test Accuracy: {test_acc:.2f}%")

    # Save Best Model
    if test_acc > best_acc:

        best_acc = test_acc

        torch.save(
            model.state_dict(),
            f"best_{MODEL_NAME}.pth"
        )

        print("Best Model Saved!")

# =========================
# Plot Accuracy & Loss
# =========================

plt.figure(figsize=(12,5))

# Accuracy Plot
plt.subplot(1,2,1)

plt.plot(
    train_accs,
    label='Train Accuracy'
)

plt.plot(
    test_accs,
    label='Test Accuracy'
)

plt.title("Accuracy Graph")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.legend()

# Loss Plot
plt.subplot(1,2,2)

plt.plot(
    train_losses,
    label='Train Loss'
)

plt.plot(
    test_losses,
    label='Test Loss'
)

plt.title("Loss Graph")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend()

plt.tight_layout()

plt.show()

# =========================
# Load Best Model
# =========================

model.load_state_dict(
    torch.load(
        f"best_{MODEL_NAME}.pth"
    )
)

# =========================
# Final Evaluation
# =========================

test_loss, test_acc, preds, labels = evaluate_model(
    model,
    test_loader
)

print(f"\nFinal Test Accuracy: {test_acc:.2f}%")

# =========================
# Classification Report
# =========================

print("\nClassification Report:\n")

print(
    classification_report(
        labels,
        preds,
        target_names=class_names
    )
)

# =========================
# Confusion Matrix
# =========================

cm = confusion_matrix(
    labels,
    preds
)

plt.figure(figsize=(8,6))

plt.imshow(
    cm,
    cmap='Blues'
)

plt.title(
    "EfficientNetV2 Confusion Matrix"
)

plt.colorbar()

plt.xlabel("Predicted Labels")

plt.ylabel("True Labels")

plt.xticks(
    np.arange(len(class_names)),
    class_names,
    rotation=45
)

plt.yticks(
    np.arange(len(class_names)),
    class_names
)

plt.show()

# =========================
# Save Final Model
# =========================

torch.save(
    model.state_dict(),
    f"final_{MODEL_NAME}.pth"
)

print("\nModel Saved Successfully!")

# =========================
# Save Class Names
# =========================

with open("class_names.txt", "w") as f:

    for item in class_names:

        f.write(f"{item}\n")

print("Class Names Saved!")
```

# Section 3: DenseNet201
**DenseNet201** connects each layer to every other layer in a feed-forward fashion, promoting feature reuse and mitigating the vanishing-gradient problem.

```python
# ============================================
# Emotion-Based Stress Classification
# Model: DenseNet201
# Dataset: FER2013 (Folder-Based)
# Platform: Kaggle Notebook
# Framework: PyTorch
# ============================================

# =========================
# Install Required Packages
# =========================


# =========================
# Import Libraries
# =========================
import os
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import (
    classification_report,
    confusion_matrix
)

import torch
import torch.nn as nn
from torch.utils.data import DataLoader

import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torchvision import models

from tqdm import tqdm

# =========================
# Device Configuration
# =========================
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print("Device:", device)

# =========================
# Dataset Paths
# =========================

TRAIN_DIR = "/kaggle/input/datasets/msambare/fer2013/train"

TEST_DIR = "/kaggle/input/datasets/msambare/fer2013/test"

# =========================
# Image Transformations
# =========================

train_transform = transforms.Compose([

    transforms.Resize((224, 224)),

    transforms.RandomHorizontalFlip(),

    transforms.RandomRotation(10),

    transforms.RandomAffine(
        degrees=10,
        translate=(0.1, 0.1)
    ),

    transforms.ColorJitter(
        brightness=0.2,
        contrast=0.2
    ),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),

    transforms.RandomErasing(
        p=0.2
    )
])

test_transform = transforms.Compose([

    transforms.Resize((224, 224)),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# =========================
# Load Datasets
# =========================

train_dataset = ImageFolder(
    root=TRAIN_DIR,
    transform=train_transform
)

test_dataset = ImageFolder(
    root=TEST_DIR,
    transform=test_transform
)

# =========================
# Class Names
# =========================

class_names = train_dataset.classes

print("Classes:", class_names)

print("Training Images:", len(train_dataset))

print("Testing Images:", len(test_dataset))

# =========================
# Visualize Sample Images
# =========================

plt.figure(figsize=(12,8))

for i in range(9):

    image, label = train_dataset[i]

    image = image.permute(1, 2, 0).numpy()

    image = np.clip(image, 0, 1)

    plt.subplot(3,3,i+1)

    plt.imshow(image)

    plt.title(class_names[label])

    plt.axis("off")

plt.tight_layout()

plt.show()

# =========================
# Data Loaders
# =========================

BATCH_SIZE = 64

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# =========================
# DenseNet201 Model
# =========================

# model = models.densenet201(
#     pretrained=True
# )

from torchvision.models import densenet201
from torchvision.models import DenseNet201_Weights

try:

    weights = DenseNet201_Weights.DEFAULT

    model = densenet201(weights=weights)

    print("Pretrained weights loaded successfully!")

except:

    print("Could not download pretrained weights.")
    print("Using model without pretrained weights.")

    model = densenet201(weights=None)

model.classifier = nn.Linear(
    model.classifier.in_features,
    7
)

model = model.to(device)

# =========================
# Loss Function
# =========================

criterion = nn.CrossEntropyLoss(
    label_smoothing=0.1
)

# =========================
# Optimizer
# =========================

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=1e-4,
    weight_decay=1e-4
)

# =========================
# Scheduler
# =========================

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    patience=2,
    factor=0.5
)

# =========================
# Training Function
# =========================

def train_model(model, loader):

    model.train()

    running_loss = 0

    correct = 0

    total = 0

    for images, labels in tqdm(loader):

        images = images.to(device)

        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)

        correct += predicted.eq(labels).sum().item()

    epoch_loss = running_loss / len(loader)

    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc

# =========================
# Evaluation Function
# =========================

def evaluate_model(model, loader):

    model.eval()

    running_loss = 0

    correct = 0

    total = 0

    predictions = []

    actuals = []

    with torch.no_grad():

        for images, labels in tqdm(loader):

            images = images.to(device)

            labels = labels.to(device)

            outputs = model(images)

            loss = criterion(outputs, labels)

            running_loss += loss.item()

            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)

            correct += predicted.eq(labels).sum().item()

            predictions.extend(
                predicted.cpu().numpy()
            )

            actuals.extend(
                labels.cpu().numpy()
            )

    epoch_loss = running_loss / len(loader)

    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc, predictions, actuals

# =========================
# Training Loop
# =========================

EPOCHS = 30

MODEL_NAME = "densenet201"

train_losses = []

test_losses = []

train_accs = []

test_accs = []

best_acc = 0

for epoch in range(EPOCHS):

    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    train_loss, train_acc = train_model(
        model,
        train_loader
    )

    test_loss, test_acc, _, _ = evaluate_model(
        model,
        test_loader
    )

    scheduler.step(test_loss)

    train_losses.append(train_loss)

    test_losses.append(test_loss)

    train_accs.append(train_acc)

    test_accs.append(test_acc)

    print(f"\nTrain Loss: {train_loss:.4f}")

    print(f"Train Accuracy: {train_acc:.2f}%")

    print(f"\nTest Loss: {test_loss:.4f}")

    print(f"Test Accuracy: {test_acc:.2f}%")

    # Save Best Model
    if test_acc > best_acc:

        best_acc = test_acc

        torch.save(
            model.state_dict(),
            f"best_{MODEL_NAME}.pth"
        )

        print("Best Model Saved!")

# =========================
# Plot Accuracy & Loss
# =========================

plt.figure(figsize=(12,5))

# Accuracy Plot
plt.subplot(1,2,1)

plt.plot(
    train_accs,
    label='Train Accuracy'
)

plt.plot(
    test_accs,
    label='Test Accuracy'
)

plt.title("Accuracy Graph")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.legend()

# Loss Plot
plt.subplot(1,2,2)

plt.plot(
    train_losses,
    label='Train Loss'
)

plt.plot(
    test_losses,
    label='Test Loss'
)

plt.title("Loss Graph")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend()

plt.tight_layout()

plt.show()

# =========================
# Load Best Model
# =========================

model.load_state_dict(
    torch.load(
        f"best_{MODEL_NAME}.pth"
    )
)

# =========================
# Final Evaluation
# =========================

test_loss, test_acc, preds, labels = evaluate_model(
    model,
    test_loader
)

print(f"\nFinal Test Accuracy: {test_acc:.2f}%")

# =========================
# Classification Report
# =========================

print("\nClassification Report:\n")

print(
    classification_report(
        labels,
        preds,
        target_names=class_names
    )
)

# =========================
# Confusion Matrix
# =========================

cm = confusion_matrix(
    labels,
    preds
)

plt.figure(figsize=(8,6))

plt.imshow(
    cm,
    cmap='Blues'
)

plt.title(
    "DenseNet201 Confusion Matrix"
)

plt.colorbar()

plt.xlabel("Predicted Labels")

plt.ylabel("True Labels")

plt.xticks(
    np.arange(len(class_names)),
    class_names,
    rotation=45
)

plt.yticks(
    np.arange(len(class_names)),
    class_names
)

plt.show()

# =========================
# Save Final Model
# =========================

torch.save(
    model.state_dict(),
    f"final_{MODEL_NAME}.pth"
)

print("\nModel Saved Successfully!")

# =========================
# Save Class Names
# =========================

with open("class_names.txt", "w") as f:

    for item in class_names:

        f.write(f"{item}\n")

print("Class Names Saved!")
```

# Section 4: MobileNetV3 Large
**MobileNetV3** represents a highly optimized, hardware-aware architecture utilizing Neural Architecture Search (NAS) and NetAdapt algorithms for efficient on-device inference.

```python
# ============================================
# Emotion-Based Stress Classification
# Model: MobileNetV3 Large
# Dataset: FER2013 (Folder-Based)
# Platform: Kaggle Notebook
# Framework: PyTorch
# ============================================

# =========================
# Import Libraries
# =========================

import os
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import (
    classification_report,
    confusion_matrix
)

import torch
import torch.nn as nn

from torch.utils.data import DataLoader

import torchvision.transforms as transforms

from torchvision.datasets import ImageFolder

from torchvision import models

from tqdm import tqdm

# =========================
# Device Configuration
# =========================

device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print("Device:", device)

# =========================
# Dataset Paths
# =========================

TRAIN_DIR = "/kaggle/input/datasets/msambare/fer2013/train"

TEST_DIR = "/kaggle/input/datasets/msambare/fer2013/test"

# =========================
# Image Transformations
# =========================

train_transform = transforms.Compose([

    transforms.Resize((224, 224)),

    transforms.RandomHorizontalFlip(),

    transforms.RandomRotation(10),

    transforms.RandomAffine(
        degrees=10,
        translate=(0.1, 0.1)
    ),

    transforms.ColorJitter(
        brightness=0.2,
        contrast=0.2
    ),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),

    transforms.RandomErasing(
        p=0.2
    )
])

test_transform = transforms.Compose([

    transforms.Resize((224, 224)),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# =========================
# Load Datasets
# =========================

train_dataset = ImageFolder(
    root=TRAIN_DIR,
    transform=train_transform
)

test_dataset = ImageFolder(
    root=TEST_DIR,
    transform=test_transform
)

# =========================
# Class Names
# =========================

class_names = train_dataset.classes

print("Classes:", class_names)

print("Training Images:", len(train_dataset))

print("Testing Images:", len(test_dataset))

# =========================
# Visualize Sample Images
# =========================

plt.figure(figsize=(12,8))

for i in range(9):

    image, label = train_dataset[i]

    image = image.permute(1, 2, 0).numpy()

    image = np.clip(image, 0, 1)

    plt.subplot(3,3,i+1)

    plt.imshow(image)

    plt.title(class_names[label])

    plt.axis("off")

plt.tight_layout()

plt.show()

# =========================
# Data Loaders
# =========================

BATCH_SIZE = 64

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# =========================
# MobileNetV3 Model
# =========================

model = models.mobilenet_v3_large(
    pretrained=True
)

model.classifier[3] = nn.Linear(
    model.classifier[3].in_features,
    7
)

model = model.to(device)

# =========================
# Loss Function
# =========================

criterion = nn.CrossEntropyLoss(
    label_smoothing=0.1
)

# =========================
# Optimizer
# =========================

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=1e-4,
    weight_decay=1e-4
)

# =========================
# Scheduler
# =========================

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    patience=2,
    factor=0.5
)

# =========================
# Training Function
# =========================

def train_model(model, loader):

    model.train()

    running_loss = 0

    correct = 0

    total = 0

    for images, labels in tqdm(loader):

        images = images.to(device)

        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)

        correct += predicted.eq(labels).sum().item()

    epoch_loss = running_loss / len(loader)

    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc

# =========================
# Evaluation Function
# =========================

def evaluate_model(model, loader):

    model.eval()

    running_loss = 0

    correct = 0

    total = 0

    predictions = []

    actuals = []

    with torch.no_grad():

        for images, labels in tqdm(loader):

            images = images.to(device)

            labels = labels.to(device)

            outputs = model(images)

            loss = criterion(outputs, labels)

            running_loss += loss.item()

            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)

            correct += predicted.eq(labels).sum().item()

            predictions.extend(
                predicted.cpu().numpy()
            )

            actuals.extend(
                labels.cpu().numpy()
            )

    epoch_loss = running_loss / len(loader)

    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc, predictions, actuals

# =========================
# Training Loop
# =========================

EPOCHS = 35

MODEL_NAME = "mobilenetv3"

train_losses = []

test_losses = []

train_accs = []

test_accs = []

best_acc = 0

for epoch in range(EPOCHS):

    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    train_loss, train_acc = train_model(
        model,
        train_loader
    )

    test_loss, test_acc, _, _ = evaluate_model(
        model,
        test_loader
    )

    scheduler.step(test_loss)

    train_losses.append(train_loss)

    test_losses.append(test_loss)

    train_accs.append(train_acc)

    test_accs.append(test_acc)

    print(f"\nTrain Loss: {train_loss:.4f}")

    print(f"Train Accuracy: {train_acc:.2f}%")

    print(f"\nTest Loss: {test_loss:.4f}")

    print(f"Test Accuracy: {test_acc:.2f}%")

    # Save Best Model
    if test_acc > best_acc:

        best_acc = test_acc

        torch.save(
            model.state_dict(),
            f"best_{MODEL_NAME}.pth"
        )

        print("Best Model Saved!")

# =========================
# Plot Accuracy & Loss
# =========================

plt.figure(figsize=(12,5))

# Accuracy Plot
plt.subplot(1,2,1)

plt.plot(
    train_accs,
    label='Train Accuracy'
)

plt.plot(
    test_accs,
    label='Test Accuracy'
)

plt.title("Accuracy Graph")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.legend()

# Loss Plot
plt.subplot(1,2,2)

plt.plot(
    train_losses,
    label='Train Loss'
)

plt.plot(
    test_losses,
    label='Test Loss'
)

plt.title("Loss Graph")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend()

plt.tight_layout()

plt.show()

# =========================
# Load Best Model
# =========================

model.load_state_dict(
    torch.load(
        f"best_{MODEL_NAME}.pth"
    )
)

# =========================
# Final Evaluation
# =========================

test_loss, test_acc, preds, labels = evaluate_model(
    model,
    test_loader
)

print(f"\nFinal Test Accuracy: {test_acc:.2f}%")

# =========================
# Classification Report
# =========================

print("\nClassification Report:\n")

print(
    classification_report(
        labels,
        preds,
        target_names=class_names
    )
)

# =========================
# Confusion Matrix
# =========================

cm = confusion_matrix(
    labels,
    preds
)

plt.figure(figsize=(8,6))

plt.imshow(
    cm,
    cmap='Blues'
)

plt.title(
    "MobileNetV3 Confusion Matrix"
)

plt.colorbar()

plt.xlabel("Predicted Labels")

plt.ylabel("True Labels")

plt.xticks(
    np.arange(len(class_names)),
    class_names,
    rotation=45
)

plt.yticks(
    np.arange(len(class_names)),
    class_names
)

plt.show()

# =========================
# Save Final Model
# =========================

torch.save(
    model.state_dict(),
    f"final_{MODEL_NAME}.pth"
)

print("\nModel Saved Successfully!")

# =========================
# Save Class Names
# =========================

with open("class_names.txt", "w") as f:

    for item in class_names:

        f.write(f"{item}\n")

print("Class Names Saved!")
```
